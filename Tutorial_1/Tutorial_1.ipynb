```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# 1. Draw 10 sample from Gaussian distribution of mean = 5 and variance = 1
mu,var = 5,1
sigma = np.sqrt(var)
constant = (1/(np.sqrt(2*(np.pi)*(sigma**2))))
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 50)
y = constant*(np.exp(-(x - mu)**2/2*(sigma**2)))
#print(x)
#print(y)

plt.title("Gaussian Distribution")
plt.xlabel("x-axis")
plt.ylabel("Probability Density")

plt.axvline(mu,linestyle = '--',label = "Mean")
plt.plot(x,y,color ='k')
plt.scatter(x,y,color ='r',label = "Sample")
plt.legend()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math
import scipy.stats as stats

# 2. Draw likelihood function for mean between 0 to 10. (Keep var = 1 constant)
mu = np.arange(0,11)
var = 1
sigma = np.sqrt(var)
constant = (1/(np.sqrt(2*(np.pi)*(sigma**2))))

x,y = [],[]
for i in mu:
  x.append(np.linspace(i - 2*sigma, i + 2*sigma, 50))
for (i,j) in zip(x,mu):
  y.append(constant*(np.exp(-(i - j)**2/2*(sigma**2))))

plt.title("Likelihood Function")
for (i,j) in zip(x,y):
  plt.plot(i,j)

xlim = [0,10]
fig, axs = plt.subplots(3,4, figsize=(30, 10))
axs = axs.ravel()

for (m,n,t) in zip(x,y,range(11)):
    axs[t].plot(m,n,color ='k')
    axs[t].axvline(t,linestyle ="--")
    axs[t].set_title("Mean = " + str(t))
    axs[t].set_xlim(xlim)
```
